```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")

        for i in self.graph[v]:
            if not visited[i]:
                self.DFSUtil(i, visited)

    def DFS(self):
        V = len(self.graph)
        visited = [False] * V

        for i in range(V):
            if not visited[i]:
                self.DFSUtil(i, visited)

g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("DFS Traversal:")
g.DFS()
```

```python
graph = {
    '5': ['3','7'],
    '3': ['2','4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = set()

def dfs(visited, graph, node):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

print("DFS Traversal:")
dfs(visited, graph, '5')
```

```python
graph = {
    'Arad': [('Zerind',75), ('Sibiu',140), ('Timisoara',118)],
    'Zerind': [('Oradea',71)],
    'Oradea': [('Sibiu',151)],
    'Timisoara': [('Lugoj',111)],
    'Lugoj': [('Mehadia',70)],
    'Mehadia': [('Drobeta',75)],
    'Drobeta': [('Craiova',120)],
    'Craiova': [('Pitesti',138), ('Rimnicu',146)],
    'Sibiu': [('Rimnicu',80), ('Fagaras',99)],
    'Rimnicu': [('Pitesti',97)],
    'Fagaras': [('Bucharest',211)],
    'Pitesti': [('Bucharest',101)],
    'Bucharest': []
}

visited = set()

def dfs(node, goal, path, cost):
    visited.add(node)
    path.append(node)

    if node == goal:
        print("Path:", " → ".join(path))
        print("Total Cost:", cost)
        return True

    for neighbor, distance in graph[node]:
        if neighbor not in visited:
            if dfs(neighbor, goal, path, cost + distance):
                return True

    path.pop()
    return False

print("DFS from Arad to Bucharest:")
dfs('Arad', 'Bucharest', [], 0)
```
