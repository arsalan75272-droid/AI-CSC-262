```python
import networkx as nx
import matplotlib.pyplot as plt

def create_ucs_graph():
    G = nx.DiGraph()

    # Add edges with weights (costs)
    edges = [
        ('S', 'A', 2),
        ('S', 'B', 5),
        ('A', 'C', 4),
        ('A', 'D', 10),
        ('B', 'D', 2),
        ('C', 'G', 3),
        ('D', 'G', 6),
        ('B', 'E', 3),
        ('E', 'G', 2)
    ]

    for u, v, w in edges:
        G.add_edge(u, v, weight=w)

    return G


def draw_graph(G):
    pos = nx.spring_layout(G, seed=42)

    # Draw nodes
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=12, font_weight='bold')

    # Draw edge labels (weights)
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    plt.title("Graph for Uniform Cost Search (UCS)")
    plt.show()


if __name__ == "__main__":
    graph = create_ucs_graph()
    draw_graph(graph)
```

```python
import heapq

def ucs_with_trace(graph, start, goal):
    frontier = [(0, start, start)]  # (cost, node, path as string)
    visited = {}

    step_label = start + "0"  # S0

    while frontier:
        cost, node, path = heapq.heappop(frontier)

        # Skip worse paths
        if node in visited and visited[node] <= cost:
            continue

        visited[node] = cost

        # Expand neighbors
        new_entries = []
        for neighbor, weight in graph[node]:
            new_cost = cost + weight
            new_path = path + neighbor
            heapq.heappush(frontier, (new_cost, neighbor, new_path))
            new_entries.append(f"{neighbor}({new_cost})")

        # Print in your desired compact style
        if new_entries:
            print(f"{step_label} {{ " + ", ".join(new_entries) + " }}")

        # Prepare next label (like A3, B5, etc.)
        step_label = f"{node}{cost}"

        # Goal check
        if node == goal:
            print("\nCost =", cost)
            print("Path =", path)
            return

    print("No solution found")


# SAME GRAPH
graph = {
    'S': [('A', 2), ('B', 5)],
    'A': [('C', 4), ('D', 10)],
    'B': [('D', 2), ('E', 3)],
    'C': [('G', 3)],
    'D': [('G', 6)],
    'E': [('G', 2)],
    'G': []
}

ucs_with_trace(graph, 'S', 'G')
```
