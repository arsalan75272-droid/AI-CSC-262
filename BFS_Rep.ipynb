```python
# Activity 1: BFS on Graph

graph = {
    '5': ['3','7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = []
queue = []

def bfs(graph, start):
    visited.append(start)
    queue.append(start)

    while queue:
        node = queue.pop(0)
        print(node, end=" ")

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

print("BFS Traversal:")
bfs(graph, '5')
```

```python
# Activity 2: BFS on Tree

def printParents(node, adj, parent):
    if parent == 0:
        print(node, "-> Root")
    else:
        print(node, "->", parent)

    for cur in adj[node]:
        if cur != parent:
            printParents(cur, adj, node)

def printChildren(Root, adj):
    q = []
    q.append(Root)

    vis = [0] * len(adj)

    while len(q) > 0:
        node = q.pop(0)
        vis[node] = 1

        print(node, "-> ", end="")
        for cur in adj[node]:
            if vis[cur] == 0:
                print(cur, end=" ")
                q.append(cur)
        print()

def printLeafNodes(Root, adj):
    for i in range(len(adj)):
        if len(adj[i]) == 1 and i != Root:
            print(i, end=" ")
    print()

def printDegrees(Root, adj):
    for i in range(1, len(adj)):
        print(i, ":", end=" ")
        if i == Root:
            print(len(adj[i]))
        else:
            print(len(adj[i]) - 1)

# Driver Code
N = 7
Root = 1

adj = [[] for _ in range(N+1)]

# Tree creation
adj[1].append(2)
adj[2].append(1)
adj[1].append(3)
adj[3].append(1)
adj[1].append(4)
adj[4].append(1)
adj[2].append(5)
adj[5].append(2)
adj[2].append(6)
adj[6].append(2)
adj[4].append(7)
adj[7].append(4)

print("Parents:")
printParents(Root, adj, 0)

print("\nChildren:")
printChildren(Root, adj)

print("Leaf Nodes:")
printLeafNodes(Root, adj)

print("Degrees:")
printDegrees(Root, adj)
```

```python
# Lab Task 1: BFS with Cost (Weighted Graph)

from collections import deque

graph = {
    'A': [('B',2), ('C',1)],
    'B': [('D',4)],
    'C': [('D',2), ('E',3)],
    'D': [('G',1)],
    'E': [('G',5)],
    'G': []
}

def bfs(graph, start, goal):
    queue = deque([(start, [start], 0)])

    while queue:
        node, path, cost = queue.popleft()

        if node == goal:
            print("Path:", path)
            print("Total Cost:", cost)
            return

        for neighbour, weight in graph[node]:
            queue.append((neighbour, path + [neighbour], cost + weight))

bfs(graph, 'A', 'G')
```

```python
# Lab Task 2: BFS for Arad to Bucharest

from collections import deque

graph = {
    'Arad': ['Sibiu', 'Timisoara'],
    'Sibiu': ['Fagaras'],
    'Timisoara': ['Lugoj'],
    'Fagaras': ['Bucharest'],
    'Lugoj': [],
    'Bucharest': []
}

def bfs(graph, start, goal):
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            print("Path:", " -> ".join(path))
            return

        for neighbour in graph[node]:
            new_path = list(path)
            new_path.append(neighbour)
            queue.append(new_path)

bfs(graph, 'Arad', 'Bucharest')
```
